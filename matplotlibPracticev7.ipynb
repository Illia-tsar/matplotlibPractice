```python
from matplotlib import pyplot as plt
from itertools import count
from matplotlib.animation import FuncAnimation
from random import randint
```

```python
#to show real time plot
%matplotlib notebook

idx = count()

x = []
y = []

plt.style.use('seaborn')

def function(i):
    x.append(next(idx))
    y.append(randint(0, 10))
    
    plt.cla()
    
    plt.plot(x, y, label='Cryptocurrency Open Price')
    
    plt.title('Open Price vs Real Time')
    plt.xlabel('Time')
    plt.ylabel('Open Price')
    
    plt.tight_layout()
    plt.legend(loc='upper left')
    
price_real_time = FuncAnimation(plt.gcf(), function, interval=500)

plt.tight_layout()
```
