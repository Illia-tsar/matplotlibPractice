```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('/home/linups/data.csv')
print(f'Data has a shape of: {data.shape}')
data.head(10)
```

```python
views = data['view_count']
likes = data['likes']
ratio = data['ratio']
```

```python
plt.style.available
```

```python
plt.style.use('fivethirtyeight')

plt.scatter(views, likes, c=ratio, cmap='Purples',
           edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Likes/Dislikes Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Views vs Likes')
plt.xlabel('Views')
plt.ylabel('Likes')

plt.tight_layout()
plt.grid(True)
```
