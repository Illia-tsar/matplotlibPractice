```python
from matplotlib import pyplot as plt
```

```python
maxTempCobar = [35.2, 28.9, 34.1, 37.6, 38.4, 41.0, 36.1, 34.0, 34.2, 35.5]
evaporationCobar = [12.0, 14.8, 12.6, 10.8, 11.4, 11.2, 13.0,  9.8, 14.6, 12.0]

maxTempRichmond = [19.6, 18.4, 20.5, 20.3, 25.2, 19.7, 23.2, 22.6, 23.0, 21.2]
evaporationRichmond = [4.3, 13.6,  1.6,  2.0,  1.0, 13.1,  1.0,  3.3,  3.4,  4.0]
```

```python
maxTempCobar.sort()
evaporationCobar.sort()
maxTempRichmond.sort()
evaporationRichmond.sort()
```

```python
with plt.xkcd():

    plt.plot(maxTempRichmond, evaporationCobar, label='Cobar')
    plt.plot(maxTempRichmond, evaporationRichmond, label='Richmond')

    plt.title('Evaporation vs MaxTemp')
    plt.xlabel('MaxTemp')
    plt.ylabel('Evaporation')

    plt.tight_layout()
    plt.grid(True)
    plt.legend()
```

```python
plt.style.use('fivethirtyeight')

plt.plot(maxTempRichmond, evaporationCobar, color='blue', linestyle='-', linewidth=1.2, marker='x', label='Cobar')
plt.plot(maxTempRichmond, evaporationRichmond, color='violet', linestyle='dotted' , marker='o', label='Richmond')

plt.title('Evaporation vs MaxTemp')
plt.xlabel('MaxTemp')
plt.ylabel('Evaporation')

plt.tight_layout()
plt.grid(True)
plt.legend(loc='lower right')
```

```python
plt.style.use('classic')

plt.plot(maxTempRichmond, evaporationCobar, color='blue', linestyle='-', linewidth=1.25,  marker='x', label='Cobar')
plt.plot(maxTempRichmond, evaporationRichmond, color='violet', linestyle='--' , linewidth=1.5, marker='o', label='Richmond')

plt.title('Evaporation vs MaxTemp')
plt.xlabel('MaxTemp')
plt.ylabel('Evaporation')

plt.grid(False)

plt.legend(loc='lower right')
```
